```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('House_Price.csv')
```

```python
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data['waterbody'].value_counts()
```

```python
data['waterbody'] = data['waterbody'].fillna('River')

# the 'waterbody' column is a categorical value, so to remove the null values takes its count and fill it with maximum count(here river has high count, so waterbody fill with river) 
```

```python
data['n_hos_beds'] = data['n_hos_beds'].fillna(data['n_hos_beds'].mean())

# the 'n_hos_beds' column is a numerical value, so it remove the null value using mean() function
```

```python
data.isnull().sum()
```

```python
le_airport = LabelEncoder()
data['airport'] = le_airport.fit_transform(data['airport'])
```

```python
le_waterbody = LabelEncoder()
data['waterbody'] = le_waterbody.fit_transform(data['waterbody'])
```

```python
le_bus_ter = LabelEncoder()
data['bus_ter'] = le_bus_ter.fit_transform(data['bus_ter'])
```

```python
data
```

```python
x = data.drop('price', axis=1)
y = data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
x_train.shape
```

### SVM have 2 hyperparameters:
* C
* Gamma

##### In C, by default it is C=1
##### In Gamma, by default it is gamma='scale

```python
model = SVR(kernel='linear') # here C=1(by default) and gamma='scale(by default)
model.fit(x_train, y_train)
print('Train score:', model.score(x_train, y_train))
print('Test score:', model.score(x_test, y_test))
```

```python
model1 = SVR(kernel='linear', C=10, gamma='auto')
model1.fit(x_train, y_train)
print('Train score:', model1.score(x_train, y_train))
print('Test score:', model1.score(x_test, y_test))
```

```python
model2 = SVR(kernel='rbf', C=20) # kernal = 'rbf' (by default) - radial basis function
model2.fit(x_train, y_train)
print('Train score:', model2.score(x_train, y_train))
print('Test score:', model2.score(x_test, y_test))
```

```python
model3 = SVR(kernel='poly', C=30)
model3.fit(x_train, y_train)
print('Train score:', model3.score(x_train, y_train))
print('Test score:', model3.score(x_test, y_test))
```
