```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('employee.csv')
```

```python
data
```

```python
data = data.sample(n=1000, random_state=34)

# to take 1000 rows from the dataset
```

```python
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
le_department = LabelEncoder()
data['department'] = le_department.fit_transform(data['department'])
```

```python
le_salary = LabelEncoder()
data['salary'] = le_salary.fit_transform(data['salary'])
```

```python
data
```

```python
data.left.value_counts()
```

```python
x = data.drop('left', axis=1)
y = data['left']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
x_train.shape
```

```python
model = SVC(kernel='linear')
model.fit(x_train, y_train)
print('Train score:', model.score(x_train, y_train))
print('Test score:', model.score(x_test, y_test))
```

```python
model1 = SVC(kernel='linear', C=10, gamma='auto')
model1.fit(x_train, y_train)
print('Train score:', model1.score(x_train, y_train))
print('Test score:', model1.score(x_test, y_test))
```

```python
model2 = SVC(kernel='rbf', C=20) 
model2.fit(x_train, y_train)
print('Train score:', model2.score(x_train, y_train))
print('Test score:', model2.score(x_test, y_test))
```

```python
model3 = SVC(kernel='poly', C=30)
model3.fit(x_train, y_train)
print('Train score:', model3.score(x_train, y_train))
print('Test score:', model3.score(x_test, y_test))
```
