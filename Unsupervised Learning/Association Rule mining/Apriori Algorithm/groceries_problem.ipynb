```python
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv('Groceries_dataset.csv')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['itemDescription'].unique()
```

```python
one_hot = pd.get_dummies(data['itemDescription'], dtype='int')
data = data.join(one_hot)
data.head()
```

```python
products = data['itemDescription'].unique()
products
```

```python
data = data.drop('itemDescription', axis=1)
```

```python
data
```

```python
records = data.groupby(['Member_number', 'Date'])[products[:]].apply(sum)
records
```

```python
def get_product_names(p):
    for product in products:
        if p[product] > 0:
            p[product] = product
    return p
records = records.apply(get_product_names, axis = 1)
```

```python
records
```

```python
records.values
```

```python
# Removing zeros

transactions = records.values
trans = []
for i in transactions:
    trans.append(i[i!=0].tolist())
trans
```

```python
rules = apriori(trans, min_support=0.0003, min_confidence=0.05, min_lift=3, min_length=2)
```

```python
rules = pd.DataFrame(rules)
```
