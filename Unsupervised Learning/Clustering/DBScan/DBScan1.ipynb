```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data with a non-linear shape (e.g., two moons)
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
# Instantiate DBSCAN model with chosen parameters
dbscan = DBSCAN(eps=0.3)
```

```python
# Fit the DBSCAN model to the data
dbscan.fit(X)
```

```python
# Retrieve the labels (cluster assignments) from DBSCAN
labels = dbscan.labels_
labels
```

```python
# Plotting the results
unique_labels = set(labels)
colors = [plt.cm.Spectral(i) for i in np.linspace(0, 1, len(unique_labels))]
```

```python
unique_labels
```

```python
colors
```

```python
plt.figure(figsize=(8, 6))
for label, color in zip(unique_labels, colors):
    if label == -1:
        # Black color for noise points (outliers)
        color = [0, 0, 0, 1]
        label_name = "Noise"
    else:
        label_name = f"Cluster {label + 1}"
    
    # Plot points with the current label
    plt.scatter(X[labels == label, 0], X[labels == label, 1], color=color, s=50, label=label_name)
    plt.title("DBSCAN Clustering on Moons Dataset")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
```
