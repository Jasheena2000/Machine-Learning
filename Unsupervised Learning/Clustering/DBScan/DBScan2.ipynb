```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# 1. Generate synthetic data using make_blobs
# Generates a dataset of 300 points with 3 centers
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# 2. Create DBSCAN model
# Set epsilon (eps) to define the neighborhood size and min_samples to define the minimum number of points for a cluster
dbscan = DBSCAN(eps=2.0, min_samples=5)

# 3. Fit the model to the data
dbscan.fit(X)

# 4. Get the cluster labels
labels = dbscan.labels_

# -1 is used for outliers/noise points
print("Cluster labels:", labels)

# 5. Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Cluster ID')
plt.legend()
plt.show()
```
